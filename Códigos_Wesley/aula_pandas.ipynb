```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("houses.xlsx")
```

```python
df.describe()
```

```python
## .unique() -  para mostrar os dados de uma coluna
dados_quartos = df["bedrooms"].unique()
print(dados_quartos)
```

```python
# mmédia dos quartos
dados_quartos[0] = 1
print(dados_quartos[0])
media_quartos = np.mean(dados_quartos)
print(media_quartos)
```

```python
# Substituindo valores nan pela média dos valores da coluna

df = df.apply(lambda col: col.fillna(col.mean()) if col.dtype in ['float64', 'int64'] else col)

# basicamente essa função está procurando valores do tipo nan em cada coluna que poderia ser feito po linhas também mas a melhor decisão é ser feita por colunas
#para substituir com o col.fillna() pela medía das colunas com col.mean() e colocando uma condição do tipo de dado se float ou inteiro, ou seja, numérico e se
# não for ele deixa do mesmo jeito como se encontra.

#   df.apply --O método apply do pandas é usado para aplicar uma função ao longo de um eixo do DataFrame (por padrão, ao longo das colunas).
#   função lambda - é uma pequena função anônima definida usando a palavra-chave lambda Uma função lambda é uma pequena função anônima definida usando a palavra-chave lambda.
#   col representa cada coluna que a função lambda vai percorrer
#   col.dtype in ['float64', 'int64']: Esta condição verifica se o tipo de dados da coluna é float64 ou int64, que são tipos numéricos.
#   col.fillna(col.mean()): Se a coluna é numérica, substitui os valores ausentes (NaN) pela média dos valores não ausentes da coluna.
#   else col: Se a coluna não é numérica (por exemplo, contém strings ou booleanos), retorna a coluna original sem alterações.
```
