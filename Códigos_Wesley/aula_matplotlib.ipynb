```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("houses.xlsx")
df = df.apply(lambda col: col.fillna(col.mean()) if col.dtype in ['float64', 'int64'] else col)
```

```python
df.head()
```

```python
df["price"].value_counts().keys()
```

```python
# fazer um gráfico entre o preço e a área quadrada de vivência
eixo_X = df["sqft_living"]
eixo_Y = df["price"]
```

```python
plt.bar(eixo_X, eixo_Y)
plt.title("Área construída X Preço")
plt.xlabel("Área construída")
plt.ylabel("Preço")
plt.grid(10)
plt.show()
```
